```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(x,y)
plt.show()
```

```python
#Training the model using Gradient Descent, I have used MSE as a LOSS Function

theta=0 #coefficient
bias=0 #intercept(bias term)
L=0.0001 #learning rate
epochs=1000 #no of itertaions the program will train itself

n=len(x) #no of elements in x

#gradient descent

for i in range(epochs): 
    y_pred = theta*x + bias #predicted value of y
    d_m = (-2/n) * sum(x*(y-y_pred)) #derivative w.r.t theta
    d_c = (-2/n) * sum(y-y_pred) #derivative w.r.t bias
    theta = theta - (L * d_m) #change in theta 
    bias = bias - (L * d_c) #change in bias
    
print(theta,bias)
```

```python
y_pred = theta*x + bias  #predicted value
plt.scatter(x,y) #plotting the given dataset
plt.scatter(x, y_pred) #plotting the predicted 
plt.show()
```

```python
plt.scatter(x, y) 
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='red')  # regression line
plt.show()
```
